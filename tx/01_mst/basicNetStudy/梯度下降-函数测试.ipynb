```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def f1(x):
    return 0.5*(x-1) **2
def f2(x):
    return x-1
```

```python
# 储存点变化位置
GD_X = []
GD_Y = []
# 其实x值
x = 21.2
# 步长
alpha = 2.1

# 初始化
f_change = f1(x)
f_current = f_change
GD_X.append(x)
GD_Y.append(f_current)
# 执行次数
iter_num = 0
BD_X = np.arange(-x,x+1,0.01)
BD_Y = np.asarray([f1(i) for i in BD_X])
```

```python
while f_change > 1e-10 and iter_num < 100:
    iter_num += 1
    x = x-alpha*f2(x)
    tmp = f1(x)
    f_change = np.abs(f_current - tmp)
    f_current = tmp
    GD_X.append(x)
    GD_Y.append(f_current)
```

```python
print(iter_num,'\t',x)
print(f'{GD_X}')
```

```python

BD_X = np.arange(-f_current,f_current+1,0.01)
BD_Y = np.asarray([f1(i) for i in BD_X])
plt.plot(BD_X,BD_Y)
plt.plot(GD_X,GD_Y)
# plt.scatter(GD_X,GD_Y)
```
